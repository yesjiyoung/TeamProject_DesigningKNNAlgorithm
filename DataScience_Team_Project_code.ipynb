```python
def info_dataset(data, verbose=True):
    label1, label2 = 0, 0
    data_size = len(data)
    for datum in data:
        if datum[-1] == 1:
            label1 += 1
        else:
            label2 += 1
    if verbose:
        print('Total of samples: %d' % data_size)
        print('Total label 1: %d' % label1)
        print('Total label 2: %d' % label2)
    return [len(data), label1, label2]
```

### Step3. split the standardized data into the train_set and the data_set.

```python
train_set, test_set = [], []
max_label1, max_label2 = int(p * label1), int(p * label2)
total_label1, total_label2 = 0, 0
for sample in standardized_data:
    if (total_label1 + total_label2) < (max_label1 + max_label2):
        train_set.append(sample)
        if sample[-1] == 1 and total_label1 < max_label1:
            total_label1 += 1
        else:
            total_label2 += 1
    else:
        test_set.append(sample)
```

### Step4. make the Manhattan_dist function ( instead of Euclidean distance )

```python
# Project Requirement Point (2) : Use Manhattan distance instead of Euclidean distance 
def manhatten_dist(p1, p2):
    dim, sum_ = len(p1), 0
    for index in range(dim - 1):
        sum_ += abs(p1[index] - p2[index])
    return sum_
```

### Step5. use the knn funtion (in the basic code)

```python
def knn(train_set, new_sample, K):
    dists, train_size = {}, len(train_set)
    
    for i in range(train_size):
        d = manhatten_dist(train_set[i], new_sample)
        dists[i] = d
    
    k_neighbors = sorted(dists, key=dists.get)[:K]
    
    qty_label1, qty_label2 = 0, 0
    for index in k_neighbors:
        if train_set[index][-1] == 1:
            qty_label1 += 1
        else:
            qty_label2 += 1
            
    if qty_label1 > qty_label2:
        return 1
    else:
        return 2
```

### Step6. to evaluate the model, set the TN, FP, FN, TN.

```python
# Define the TN, FP, FN, TN
#TP : real 1 - knn 1
#FP : real 2 - knn 2
#FN : real 2 - knn 1
#TN : real 1 - knn 2
```

```python
TN= 0
TP= 0
K = 3
for sample in test_set:
    label = knn(train_set, sample, K) # test값 
    if sample[-1] == label: # real이랑 test값 비교 
        if label == 1 : 
             TP += 1
        else :
            TN += 1
```

```python
FP= 0
FN= 0
K = 3
for sample in test_set:
    label = knn(train_set, sample, K)
    if sample[-1] != label: 
        if label == 1 :
            FP += 1
        else :
            FN += 1
```

### Step7. calculate and show Accuracy, Precision and Recall 

```python
# Project Requirement Point (3) : Calculate and Show Precision and Recall
Accuracy = (TN+TP)/(TN+FP+FN+TP)
Precision = TP / (FP+TP)
Recall= TP / (FN+TP)
```
